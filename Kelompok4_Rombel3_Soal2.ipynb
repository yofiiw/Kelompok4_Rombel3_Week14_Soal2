```python
# Kelompok 4 Soal nomor 2 UAS 2024 Semester 1 Rombel 3
class Buku:
    def __init__(self, judul, pengarang, tahun_terbit, jumlah_stok):             # Zulfa Nashihin 2420506036
        self.judul = judul                                                       # Yofi Widiyanto 2420506039
        self.pengarang = pengarang                                               # Novela Eka Saputri 2420506040
        self.tahun_terbit = tahun_terbit
        self.jumlah_stok = jumlah_stok

    def pinjam(self):
        if self.jumlah_stok > 0:
            self.jumlah_stok -= 1
            print(f"Buku '{self.judul}' berhasil dipinjam.")
        else:
            print(f"Maaf, stok buku '{self.judul}' habis.")

    def kembalikan(self):
        self.jumlah_stok += 1
        print(f"Buku '{self.judul}' berhasil dikembalikan.")


# List dan dictionary untuk menyimpan data buku
daftar_buku = []
dict_buku = {}


# Fungsi untuk menampilkan semua buku
def tampilkan_semua_buku():
    if not daftar_buku:
        print("Belum ada buku yang tersedia.")
    else:
        print("\nDaftar Buku:")
        for buku in daftar_buku:
            print(f"Judul: {buku.judul}, Pengarang: {buku.pengarang}, Tahun: {buku.tahun_terbit}, Stok: {buku.jumlah_stok}")


# Fungsi untuk mencari buku berdasarkan judul
def cari_buku(judul):
    for buku in daftar_buku:
        if buku.judul.lower() == judul.lower():
            print(f"Buku ditemukan: Judul: {buku.judul}, Pengarang: {buku.pengarang}, Tahun: {buku.tahun_terbit}, Stok: {buku.jumlah_stok}")
            return buku
    print(f"Buku dengan judul '{judul}' tidak ditemukan.")
    return None


# Fungsi untuk menambahkan buku baru
def tambah_buku():
    try:
        judul = input("Masukkan judul buku: ")
        pengarang = input("Masukkan pengarang buku: ")
        if not judul or not pengarang:
            raise ValueError("Judul dan pengarang harus diisi.")

        # Memastikan tahun terbit valid
        tahun_terbit = input("Masukkan tahun terbit buku: ")
        if not tahun_terbit.isdigit() or len(tahun_terbit) < 3:
            raise ValueError("Tahun terbit harus berupa angka dengan minimal 3 digit.")
        tahun_terbit = int(tahun_terbit)

        # Memastikan jumlah stok valid
        jumlah_stok = input("Masukkan jumlah stok buku: ")  # Take input as string
        if not jumlah_stok.isdigit():  # Check if string is digit and then convert to int for comparison
            raise ValueError("Stok yang dimasukkan harus berupa angka")
        elif int(jumlah_stok) < 1:  # Convert to int after validation
            raise ValueError("Stok yang dimasukkan harus lebih dari 0")
        jumlah_stok = int(jumlah_stok)

        buku_baru = Buku(judul, pengarang, tahun_terbit, jumlah_stok)
        daftar_buku.append(buku_baru)
        dict_buku[judul] = buku_baru
        print(f"Buku '{judul}' berhasil ditambahkan.")
    except ValueError as e:
        print(f"Input tidak valid: {e}")


# Fungsi utama program
def menu():
    while True:
        print("\n=== Menu Manajemen Perpustakaan ===")
        print("1. Tambah Buku")
        print("2. Tampilkan Semua Buku")
        print("3. Cari Buku")
        print("4. Pinjam Buku")
        print("5. Kembalikan Buku")
        print("6. Keluar")
        pilihan = input("Pilih menu: ")

        if pilihan == "1":
            tambah_buku()
        elif pilihan == "2":
            tampilkan_semua_buku()
        elif pilihan == "3":
            judul = input("Masukkan judul buku yang dicari: ")
            cari_buku(judul)
        elif pilihan == "4":
            judul = input("Masukkan judul buku yang ingin dipinjam: ")
            buku = cari_buku(judul)
            if buku:
                buku.pinjam()
        elif pilihan == "5":
            judul = input("Masukkan judul buku yang ingin dikembalikan: ")
            buku = cari_buku(judul)
            if buku:
                buku.kembalikan()
        elif pilihan == "6":
            print("Terima kasih telah menggunakan program ini.")
            break
        else:
            print("Pilihan tidak valid. Silakan coba lagi.")


# Menjalankan program
if __name__ == "__main__":
    menu()
```
